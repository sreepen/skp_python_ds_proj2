# Loading the dataset: 

This kaggle dataset is of popular Spotify tracks with over 60,000 songs. It has popular songs from 6 different languages such as English, Hindi, Tamil, Telugu (my mother tongue!), Malyalam and Korean. The dataset has several columns some of which include the track name, the year of release, the popularity of the song, the name of the album, the language, and the tempo to name a few. 

```python
import pandas as pd
spotify_data = pd.read_csv('spotify_tracks.csv')

#this is just for a check:
print(spotify_data.head())
print(spotify_data.describe())
```

# Mapping the questions as queries:
**To make it easier to understand the data, I thought it would be best to represent the answers to these quesries in the form of graphs, plots, etc.**

(In order to do so, I will be using some of the concepts I have learned from my DS 200 class when it comes to python coding, jupyter notebooks, plotting data, etc.)

```python
import matplotlib.pyplot as plt
```

**1. In which year were most tracks released?**

```python
releases_per_year = spotify_data['year'].value_counts().sort_index()

plt.figure(figsize=(12,6))
releases_per_year.plot(kind = 'bar', color = 'blue')
plt.title('Number of Track Releases Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Releases')

plt.show()
```

**2. Top 10 artists with most tracks in this dataset?**

```python
top_artists = spotify_data['artist_name'].value_counts().head(10)

plt.figure(figsize=(10,6))
top_artists.plot(kind = 'bar', color = 'blue')
plt.title('Top 10 Artists with Most Tracks')
plt.xlabel('Artist')
plt.ylabel('Number of Tracks')

plt.show()
```

**3. What is the average duration of tracks over time?**

```python
avg_duration = spotify_data.groupby('year')['duration_ms'].mean()

plt.figure(figsize=(12,6))
avg_duration.plot(kind='line', color = 'blue')
plt.title('Avrage Duration of Tracks Over the Years')
plt.xlabel('Year')
plt.ylabel('Duration (in milliseconds)')

plt.show()
```

**4. Which language is the most dominant in this dataset?**

```python
top_lang = spotify_data['language'].value_counts().head(6)

plt.figure(figsize=(10,6))
top_lang.plot(kind = 'bar', color = 'blue')
plt.title('Top languages in Dataset')
plt.xlabel('Language')
plt.ylabel('Count')

plt.show()
```

**5. What is the distribution of track popularity? (Are there more popular tracks or less popular tracks in this dataset?)**

```python
import seaborn as sns

plt.figure(figsize=(10,6))
sns.histplot(spotify_data['popularity'], kde = True, bins = 20, color = 'blue')
plt.title('Distribution of Track Popularity')
plt.xlabel('Popularity')
plt.ylabel('Frequency')

plt.show()
```

**6. Does the danceability of a song have an impact on its popularity?**

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data = spotify_data, x = 'danceability', y = 'popularity', size = 'popularity', sizes = (10, 100))
plt.title('Danceability vs Popularity')
plt.xlabel('Danceability')
plt.ylabel('Popularity')

plt.show()
```
